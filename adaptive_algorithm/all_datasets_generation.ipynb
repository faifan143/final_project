# Sudoku

```python
import random
import csv

# Function to generate completion time
def generate_completion_time():
    return round(random.uniform(5, 60), 2)  # Generates completion time between 5 to 60 minutes

# Function to generate hints
def generate_hints():
    return random.randint(0, 3)  # Generates hints out of 3

# Function to generate mistakes
def generate_mistakes():
    return random.randint(0, 3)  # Generates mistakes out of 3

# Function to generate is_completed
def generate_is_completed():
    return random.randint(0,1)  # Generates is_completed on a scale from 0 to 1

# Function to generate previous level
def generate_previous_level():
    return random.randint(1, 4)  # Generates previous level on a scale from 1 to 4

# Function to generate new level based on stats
def generate_new_level(completion_time, hints, mistakes, is_completed, previous_level):
    if completion_time <= 10 and hints == 0 and mistakes == 0 and is_completed == 1:
        return previous_level + 1 if previous_level < 4 else 4
    elif completion_time <= 10 and (hints > 0 or mistakes > 0) and is_completed == 1:
         return previous_level
    else :
        return previous_level - 1 if previous_level > 1 else 1


# Generate dataset
dataset = []
for _ in range(600000):
    completion_time = generate_completion_time()
    hints = generate_hints()
    mistakes = generate_mistakes()
    is_completed = generate_is_completed()
    previous_level = generate_previous_level()
    new_level = generate_new_level(completion_time, hints, mistakes, is_completed, previous_level)
    dataset.append({
        'completion_time': completion_time,
        'hints': hints,
        'mistakes': mistakes,
        'is_completed': is_completed,
        'previous_level': previous_level,
        'new_level': new_level
    })


    # Generate dataset
new_dataset = []
for _ in range(300000):
    completion_time = round(random.uniform(5, 10), 2)
    hints =0
    mistakes = 0
    is_completed = 1
    previous_level = generate_previous_level()
    new_level = generate_new_level(completion_time, hints, mistakes, is_completed, previous_level)
    new_dataset.append({
        'completion_time': completion_time,
        'hints': hints,
        'mistakes': mistakes,
        'is_completed': is_completed,
        'previous_level': previous_level,
        'new_level': new_level
    })

dataset.extend(new_dataset)

      
new_dataset = []
for _ in range(100000):
    completion_time = round(random.uniform(5, 10), 2)
    hints =random.randint(1, 3)
    mistakes = random.randint(1, 3)
    is_completed = 1
    previous_level = generate_previous_level()
    new_level = generate_new_level(completion_time, hints, mistakes, is_completed, previous_level)
    new_dataset.append({
        'completion_time': completion_time,
        'hints': hints,
        'mistakes': mistakes,
        'is_completed': is_completed,
        'previous_level': previous_level,
        'new_level': new_level
    })

dataset.extend(new_dataset)

csv_file = 'datasets/sudoku_players_stats.csv'

random.shuffle(dataset)

df = pd.DataFrame(dataset)

df.to_csv(csv_file , index=False )

```

# Connect4

```python
import random

player_stats = []
for _ in range(1000000):
    completion_time = random.uniform(1, 10)  # Completion time in minutes
    is_completed = random.choice([0, 1])     # 0 or 1 indicating whether the game is completed
    hints = 0                                # Hints always 0
    mistakes = 0                             # Mistakes always 0
    previous_level = random.randint(1, 4)    # Previous level on a scale from 1 to 4

    if completion_time <= 5 and is_completed == 1:
        new_level = min(previous_level + 1, 4)
    elif completion_time > 5 and completion_time <= 7.5:
        new_level = previous_level
    else:
        new_level = max(previous_level - 1, 1)

    player_stats.append({
        'completion_time': round(completion_time, 2),
        'hints': hints,
        'mistakes': mistakes,
        'is_completed': is_completed,
        'previous_level': previous_level,
        'new_level': new_level
    })

random.shuffle(player_stats)

df = pd.DataFrame(player_stats)

df.to_csv('datasets/connect4_players_stats.csv' , index=False )
```

# Shuffler

```python
import csv
import random

datasets = [
    'datasets/chess_players_stats.csv',
    'datasets/connect4_players_stats.csv',
    'datasets/dots_boxes_players_stats.csv',
    'datasets/jigsaw_players_stats.csv',
    'datasets/memory_cards_players_stats.csv',
    'datasets/missing_words_players_stats.csv',
    'datasets/sudoku_players_stats.csv',
]

def shuffle_csv(filename):
    # Read the CSV file and store its rows in a list
    with open(filename, 'r', newline='') as file:
        reader = csv.reader(file)
        rows = list(reader)
    
    # Shuffle the rows
    random.shuffle(rows)
    
    # Write the shuffled rows back to the CSV file
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)
        writer.writerows(rows)

# Shuffle each dataset
for dataset in datasets:
    shuffle_csv(dataset)
```
